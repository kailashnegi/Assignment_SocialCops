<b>We'll use PCA technique to find the score of each district based on these variables so that we can rank them.</b>

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

from sklearn.decomposition import PCA
```

```python
dist_level_df = pd.read_csv("C:/Users/abc/Desktop/Data Analytics/Projects/SocialCops - Assignment/Distric_Level_Data.csv",index_col=0)
dist_level_df.index.name = None
dist_level_df.head()
```

```python
#Correlation among variables
corr = dist_level_df.corr()
corr
```

```python
#Normalization
dist_level_df_norm = (dist_level_df - dist_level_df.mean()) / dist_level_df.std()
```

```python
dist_level_df_norm.head()
```

```python
##PCA Model Fitting
pca = PCA(n_components=15)
pca.fit(dist_level_df_norm)
```

```python
# PCA Loadings
loading = pca.components_
loading
```

```python
##First Principal Component factorscore
score = loading[0]
```

```python
# Summing all the normalized values of the dataframe after multiplying with the corresponding score
dist_level_df_norm['Score'] = (dist_level_df_norm.multiply(score)).sum(axis=1)
```

```python
##Sorting districts based on Scores
dist_level_df_norm = dist_level_df_norm.sort(columns='Score', ascending = False)
```

```python
##Here is the ranking of districts based on Scores.
print(dist_level_df_norm['Score'])
```

```python
# Exporting districts with their score or ranking in an excel file
dist_level_df_norm.to_csv('C:/Users/abc/Desktop/Data Analytics/Projects/SocialCops - Assignment/Index_Ranking.csv', header= True, columns = {'Score'})
```
